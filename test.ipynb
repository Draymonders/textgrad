```python
import sys
sys.path.append('../')
import os
import textgrad as tg
from textgrad.engine.openai import ChatOpenAI
```

```python
os.environ["OPENAI_API_KEY"] = ""
model = ""
url = ""

engine = ChatOpenAI(model_string=model, base_url=url)
model = tg.BlackboxLLM(engine=engine)

tg.set_backward_engine(engine)
question_string = ("2025年4月黄金行情波动原因？")
question = tg.Variable(question_string,
                       role_description="question to the LLM",
                       requires_grad=False)
# print(question)
answer = model(question)
```

```python
answer.set_role_description("对问题简洁准确的回答")

# Step 2: Define the loss function and the optimizer, just like in PyTorch!
# Here, we don't have SGD, but we have TGD (Textual Gradient Descent)
# that works with "textual gradients".
evaluation_instruction = (f"这是问题: {question_string}. "
                           "\n已知的是黄金在4月暴涨后又暴跌了。并且美国的关税政策在不断摇摆 "
                           "\n评估对这个问题的任何给定答案：要聪明、有逻辑且极具批判性。只需提供简洁的反馈。")
loss_fn = tg.TextLoss(evaluation_instruction)
loss = loss_fn(answer)
# print(loss)
```

```python
# TextLoss is a natural-language specified loss function that describes
# how we want to evaluate the reasoning.
optimizer = tg.TGD(parameters=[answer])
loss.backward()
# print(answer)
optimizer.step()
# print(answer)
```
