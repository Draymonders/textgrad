```python
import sys
sys.path.append('../')
import os
import textgrad as tg
from textgrad.engine.openai import ChatOpenAI

#4f16fa15-dc20-4a33-8443-a13b08e418db
os.environ["OPENAI_API_KEY"] = ""
model = ""
url = ""

engine = ChatOpenAI(model_string=model, base_url=url)
model = tg.BlackboxLLM(engine=engine)
tg.set_backward_engine(engine)
```

```python
# Define a variable to optimize
solution = tg.Variable("简要介绍下马斯克，不超过200字", 
                      requires_grad=True,
                      role_description="人物介绍")

# Generate a response
response = model(solution)

# Define a loss function
loss_fn = tg.TextLoss("说明重要经历及其年限，包括其与特朗普的关系")

# Compute loss and gradients
loss = loss_fn(response) # 计算loss
loss.backward()

# Create optimizer and update the variable
optimizer = tg.TGD(parameters=[solution])
optimizer.step()

# The solution variable now contains improved text
print(solution.value)
```
